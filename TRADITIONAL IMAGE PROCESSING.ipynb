# TRADITIONAL IMAGE PROCESSING

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#1st Image
path1=r"C:\Users\User\Downloads\Dataset\zebra long wing\080.jpg"
image1=cv2.imread(path1)
cv2.imshow("Original Image",image1)
cv2.waitKey(0)
cv2.destroyAllWindows

#2nd Image
path2=r"C:\Users\User\Downloads\Dataset\wood satyr\067.jpg"
image2=cv2.imread(path2)
cv2.imshow("Original Image1",image2)
cv2.waitKey(0)
cv2.destroyAllWindows
```

```python
image1.shape
image2.shape
```

```python
b1,g1,r1 =cv2.split(image1)
b2,g2,r2 =cv2.split(image2)
```

```python
# Plot histogram for R channel of image1
plt.hist(r1.ravel(),256,[0,256], color='red', alpha=0.5, label='Image 1')

# Plot histogram for R channel of image2
plt.hist(r2.ravel(),256,[0,256], color='blue', alpha=0.5, label='Image 2')

# Add legend and labels
plt.legend()
plt.title('Histogram of R channel')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# Plot histogram for G channel of image1
plt.hist(g1.ravel(),256,[0,256], color='red', alpha=0.5, label='Image 1')

# Plot histogram for G channel of image2
plt.hist(g2.ravel(),256,[0,256], color='blue', alpha=0.5, label='Image 2')

# Add legend and labels
plt.legend()
plt.title('Histogram of G channel')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# Plot histogram for B channel of image1
plt.hist(b1.ravel(),256,[0,256], color='red', alpha=0.5, label='Image 1')

# Plot histogram for B channel of image2
plt.hist(b2.ravel(),256,[0,256], color='blue', alpha=0.5, label='Image 2')

# Add legend and labels
plt.legend()
plt.title('Histogram of B channel')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
image1_hsv = cv2.cvtColor(image1, cv2.COLOR_BGR2HSV)
image2_hsv = cv2.cvtColor(image2, cv2.COLOR_BGR2HSV)

h1, s1, v1 = cv2.split(image1_hsv)
h2, s2, v2 = cv2.split(image2_hsv)
```

```python
# Plot histogram for HUE of image1
plt.hist(h1.ravel(),256,[0,256], color='red', alpha=0.5, label='Image 1')

# Plot histogram for HUE of image2
plt.hist(h2.ravel(),256,[0,256], color='blue', alpha=0.5, label='Image 2')

# Add legend and labels
plt.legend()
plt.title('HUE')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# Plot histogram for SATURATION of image1
plt.hist(s1.ravel(),256,[0,256], color='red', alpha=0.5, label='Image 1')

# Plot histogram for SATURATION  of image2
plt.hist(s2.ravel(),256,[0,256], color='blue', alpha=0.5, label='Image 2')

# Add legend and labels
plt.legend()
plt.title('SATURATION ')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# Plot histogram for VALUE of image1
plt.hist(v1.ravel(),256,[0,256], color='red', alpha=0.5, label='Image 1')

# Plot histogram for VALUE of image2
plt.hist(v2.ravel(),256,[0,256], color='blue', alpha=0.5, label='Image 2')

# Add legend and labels
plt.legend()
plt.title('VALUE ')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```
